```python
#ライブラリのインポート
from time import sleep
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# 変数urlにスクレイピングしたいURLを格納する
url = 'https://doga.hikakujoho.com/netflix/japanesedrama/?page={}'

#変数outputに空のリストを作成する
output = []

#現在のurlを表示
print(url)
```

```python
#ループさせる
for i in range(1,3):
    
    #ループさせるために、target_urlの作成
    target_url = url.format(i)

    #dlistの長さを確認
    print("outputの大きさ:",len(output))
    
    #print()をしてtarget_urlを確かめる
    print(target_url)
 

    ##ここから実装が必要
    #ループさせるために、target_urlの作成

    #HTMLを取得
    res = requests.get(target_url)
    
    #スリープの実装
    sleep(2)

    #取得したHTMLを'BeautifulSoup'で解析
    soup = BeautifulSoup(res.text, 'html.parser')

    list_show_update = soup.find('ul',class_='list_show_update')

    span_tag = list_show_update.find_all('span',class_='fwb')

    span_strings = [x.string for x in span_tag]

    output.append(span_strings)
    
    print(output)
    
```

```python
##データフレームを作成する
df=pd.DataFrame(output)

#中身の確認
df.head()
```

```python
##to_csvを使用してcsv出力する
df.to_csv('NETFLIX_国内ドラマ.csv' ,encoding = 'utf-8-sig')
```
